```python
## How to handle or replace the null values?
## Should choose mean or median? Usually, we choose the median over the mean 
## Because the mean gets affected by the outliers and may skew the distribution of overall data
## Whereas Median is robust to outliers
## Now again we won't fill all the values with a single/same median, it would be bias since we have multiple categories
## We can't put one in all the shoes, clothing, etc. 
## So we will impute the median/values of review ratings within each category
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating']. transform(lambda x: x.fillna(x.median())) 
```

```python
df.isnull().sum()
```

```python
## Let's organise the column names, good practice to have everything in lowercase
## it makes our code consistent, easier to read, and most imp it easy to reference the code in SQL, python
## Without runnin ginto erros.

df.columns = df.columns.str.lower()   #firstly we are changing everything to lower case
df.columns = df.columns.str.replace(' ','_') #secondly, we are replacing the spaces with underscore
```

```python
df.columns
```

```python
# create a column age_group (doing this to group customers into 4 groups
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)

# qcut: It splits the data into 4 equal sized groups based on data distribution and assign the given labels
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days
# right now the column that we have is how often the purchase has been made in weekly and monthly format
# having sentence type makes the analysis harder, numbers are easy to compare and good to have for analysis

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quaterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
## do we actually need both these columns? or Is this redundancy? lets check!

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
## hence proved that both are carrying the same information, lets remove one i.e. promo_code_used

df = df.drop('promo_code_used', axis=1)
```
